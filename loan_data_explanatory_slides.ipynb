## Investigation Overview

Loans and credit is ever present in most of our lives. Along with credit comes the interest that we pay to make use of the credit. In this investigation, I would like to understand and explore what are the parameters based on which the interest for the credit extended is dependant on. Specifically, we will be looking at the Annual Percentage Rate (APR) for the loan.

## Dataset Overview

The loan data set provided by prosper contains data for upwards of 110K loans listings, with 81 variables about each loan such as loan amount, interest rate, loan status, borrower income and so on. 

Since we have 81 variables that describes each loan and it may not be humanly possible to identify the relations between all combinations of these variables, I have used an educated guess to determine what parameters could potentially be related to calculation of interest before disbursal of the loan.

I have also removed some data that was missing or wasn't useful such as NC and NaN `Credit Rating`, `Credit Scores` of 0, and the key variable of interest - `BorrowerAPR` where the data was NaN.

```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")

base_color = sb.color_palette()[0]
```

```python
# load in the dataset into a pandas dataframe
loans = pd.read_csv('prosperLoanData_filtered.csv')
loans.head()
```

```python
# check data types
loans.info()
```

```python
# Provide all the ordinal data
categories = {'CreditRating': ['HR', 'E', 'D', 'C', 'B', 'A','AA'],
              'LoanOriginationQuarter': ['Q1', 'Q2', 'Q3', 'Q4'],
              'EmploymentStatus': ['Not available', 'Not employed', 'Retired', 'Part-time', 'Other', 'Self-employed', 'Employed', 'Full-time']}

for varname, varlist in categories.items():
    vartype = pd.api.types.CategoricalDtype(varlist, ordered=True)
    loans.loc[:,varname] = loans[varname].astype(vartype)
```

> Note that the above cells have been set as "Skip"-type slides. That means
that when the notebook is rendered as http slides, those cells won't show up.

## Loans are ever rising!

- About 34000 loans were taken in 2013 amounting to about __350 Million USD__.

> Credit is an integral part of our lives. To satisfy our desires we may have to go beyond our current savings and liquidity and indulge in credit. The amount of credit we have been taking up, which had been steady for a few years, has been increasing exponentially Year on Year<sup>(1)</sup> since 2009.<sup>(2)</sup>


<sub>(1) data is partial in 2014 and has been collected only until March 2014.</sub><br>
<sub>(2) 2009 saw a lull in lending, probably as a result of recession and lending markets being down.</sub>

```python
plt.figure(figsize=(15,5))

plt.subplot(1,2,1)
sb.countplot(data=loans, x='LoanOriginationYear', color=base_color)
plt.title('Yearly Number of Loans taken')
plt.xlabel('Year')
plt.ylabel('Number of loans taken')

plt.subplot(1,2,2)
sb.barplot(data=loans, x='LoanOriginationYear', y='LoanOriginalAmount', color=base_color, estimator=np.sum, ci=None);
plt.title('Yearly Amount of Loans taken')
plt.xlabel('Year')
plt.ylabel('Amount of loans taken (in Million USD)')
plt.ticklabel_format(axis='y', scilimits=(6,6))

plt.show();
```

## Increasing Outflow due to Loans
 
- An individual has 10 current credit lines on an average.
> That is a lot of credit that each person is having to manage and repay.

- Average borrower APR<sup>(1)</sup> is 36%.<sup>(2)</sup>
>When a credit is taken, it is not just the amount taken that has to be repaid, but the borrower is charged an interest on top of this amount that they have to repay. This is an additional outflow from the income and this only increases with increasing credit lines.

This APR seems to fluctuate over a large range and the borrower might be interested in knowing what affects their APR when they are looking for new credit.

<sub>(1) APR is the `Annual Percentage Rate` of interest on loans.</sub><br>
<sub>(2) APR in the data is represented in fractional values.</sub>

```python
plt.figure(figsize=(15,5))

plt.subplot(1,2,1)
bin_edges = np.arange(0,0.55,0.01)
plt.hist(loans['BorrowerAPR'], bins=bin_edges)
plt.xlim(0,0.45)

plt.title('Distribution of Borrower APR')
plt.xlabel('Borrower APR')
plt.ylabel('Frequency')

plt.subplot(1,2,2)
bin_edges = np.arange(0,55,1)
plt.hist(loans.loc[:, 'CurrentCreditLines'], bins = bin_edges)

plt.title('Distribution of credit lines')
plt.xlabel('Current Credit Lines')
plt.ylabel('Frequency')
plt.xlim(0,35)

plt.show();
```

### Factors affecting Borrower APR

- Borrower APR seems to most correlated to the following factors
  - Loan Amount taken
  - Credit Score
  
> In both cases, we see a negative correlation. As the Loan Amount increases, the upper range of Borrower APR seems to come down. To a much higher extent, the APR is correlated (negatively) to Credit Scores; the Borrower APR reduces as the Credit Score increases.

```python
plt.figure(figsize=(15,7))
loans_sample = loans.sample(10000)

plt.subplot(1,2,1)
plt.scatter(y=loans_sample['BorrowerAPR'], x=loans_sample['LoanOriginalAmount'], alpha=0.3)
plt.xscale('log')
xticks = [1000,1500,2000,3000,4000,5000,7000,10000,15000,20000,25000,30000]
xlabels=['1','1.5','2','3','4','5','7','10','15','20','','30']
plt.xticks(xticks, xlabels)

plt.ylabel('Borrower APR')
plt.xlabel('Average Loan Amount (in 1000 USD)')
plt.title('Relation of Borrower APR to Loan Amount')
plt.ylim(0.05,.4)

plt.subplot(1,2,2)
plt.scatter(y=loans_sample['BorrowerAPR'], x=loans_sample['CreditScoreRangeAverage'], alpha=0.3)
plt.ylabel('Borrower APR')
plt.xlabel('Average Credit Score')
plt.title('Relation of Borrower APR to Credit Scores')
plt.ylim(.05,.4)

plt.show();
```

### Market Forces affecting APR

- Market forces are also showing a trend in the overall borrowing rates. 
> The interest rates have been on an upward trend until 2011 after which we see it sharply coming back down. 

- 2014 has seen the lowest Interest Rates.
> No wonder then that there was never a better time than now (2013-2014) for getting credit at low rates, and this shows by the increasing amount of credit that is being taken.

```python
plt.figure(figsize=(10,5))
sb.boxplot(data=loans, x='LoanOriginationYear', y='BorrowerAPR', palette='Greens');

plt.title('Distribution of Borrower APR over the years')
plt.xlabel('Loan Origination Year')
plt.ylabel('Borrower APR')
plt.show();
```

### The relevance of Credit Ratings

- We see how strongly correlated Credit Ratings are to the Credit Scores.
> Credit Ratings are given to individual based on factors such as credit scores, existing credit lines, repayment practices, and so on. In fact Prosper seems to be using a special combination of signals to come up with a Credit Rating for each individual.

- We also see that Borrower APR is very strongly (negatively) correlated to the Credit Ratings.
> Clearly APR reduces with better Credit Ratings. The mean APR is reducing linearly, as well as the Inter Quartile spread of APR is getting narrower emphasising the downward trend.

```python
base_color = sb.color_palette()[0]
plt.figure(figsize=(15,7))

plt.subplot(1,2,1)
sb.boxplot(data=loans, x='CreditRating', y='CreditScoreRangeAverage', color=base_color)
plt.xlabel('Credit Rating')
plt.ylabel('Average Credit Score')
plt.title('Relation of Credit Score to Credit Rating')


plt.subplot(1,2,2)
sb.boxplot(data=loans, x='CreditRating', y='BorrowerAPR', palette='Greens')
plt.xlabel('Credit Rating')
plt.ylabel('Borrower APR')
plt.title('Relation of Borrower APR to Credit Ratings')

plt.show();
```

### Distributions of Credit Ratings

- Most individuals are in D-C-B band of Credit Ratings.
>We see that Credit Ratings are normally distributed across the population, and most of the individuals are in the D-C-B band of Credit Ratings. 

- Most loans are taken by individuals with Credit Rating C.
>In fact, looking at the Amount of Loans taken across the Credit Ratings, we can see a similar distribution where the maximum amount of loans taken<sup>(1)</sup> are peaking at C.

<sub>(1) Interestingly, higher credit rating correlates with more loans, and increased loan amount eligibility.</sub>

```python
base_color = sb.color_palette()[0]
plt.figure(figsize=(15,7))

plt.subplot(1,2,1)
sb.countplot(data=loans, x='CreditRating', color=base_color)
plt.xlabel('Credit Rating')
plt.ylabel('Frequency')
plt.title('Distribution of Credit Ratings')

plt.subplot(1,2,2)
sb.barplot(data=loans, x='CreditRating', y='LoanOriginalAmount', estimator=np.sum, color=base_color, ci=None);
plt.title('Relation of Credit Ratings to Loans Taken')
plt.xlabel('Credit Rating')
plt.ylabel('Total Loan Amount (in Million USD)')
plt.ticklabel_format(axis='y', scilimits=(6,6))

plt.show();
```

### APR as a function of Credit Rating and Loan Amount

- Credit Ratings seem to be a key factor to determine APR. 
> By looking at how the APR changes as a function of both loan amount and credit rating <sup>(1)</sup>, very clear trends start to emerge. For one, we can see that Credit Ratings start showing a clear band for values of APR. This shows the very strong correlation between the two.

- Borrower APR range narrows with increasing Loan Amount.
> The triangular distribution is clearly showing how the interest range has a wide spread at lower loan amounts, where at higher loan amounts, the APR doesn't fluctuate much.

- Loan amount sizes are related to Credit Rating.
> The trend is mostly normal, but its clear how the sizes of loans taken are very clearly correlated to the Credit Rating. It seems like at lower Credit Ratings, the eligibility to take larger loans is not there. Whereas at higher Credit Ratings, the necessity to take larger loans is not there (and hence better credit scores).


<sub>(1) data is limited to the year 2013 to show the clear trend.</sub><br>

```python
loans_2013 = loans.query('LoanOriginationYear == 2013')

plt.figure(figsize=(15,8))
sb.scatterplot(data=loans_2013, x='LoanOriginalAmount', y='BorrowerAPR', hue='CreditRating', palette='cubehelix_r')

plt.title('Relation of Borrower APR with Loan Amount and Credit Ratings (in 2013)')
plt.xlabel('Loan Amount (in USD)')
plt.ylabel('Borrower APR')
plt.show();
```

### Monthly Incomes, Credit Lines, and Loan Amounts

Why are we focussing so much on Loan Amounts and not other the other monetary factors like Income?

- More the Income, more the Loan Amount.
> It goes without saying that with an increase in monthly income, the eligibility or appetite to take larger loans only increases.

- Increasing Credit Lines with Income.
> Within a given loan amount size, it can be seen that the number of Credit Lines tends to incease with increase in Incomes. 

All this goes to show that the 3 features are closely related and hence it makes sense to consider just one of them in order to observe the relevant patterns. Loan Amounts are where the patterns come out very clearly.

```python
plt.figure(figsize=(15,10))
loans_sample = loans.query('CurrentCreditLines <= 30')
loans_sample = loans_sample.sample(10000)
sb.scatterplot(data=loans_sample, y='StatedMonthlyIncome', x='LoanOriginalAmount', 
               hue='CurrentCreditLines', size='CurrentCreditLines', sizes=(5,200), palette='Blues')
plt.xscale('log')
xticks = [1000,1500,2000,3000,4000,5000,7000,10000,15000,20000,25000,30000]
xlabels=['1','1.5','2','3','4','5','7','10','15','20','','30']
plt.xticks(xticks, xlabels)

plt.ylim(500,15000)

plt.title('Relation of Loan Amount to Monthly Income and Credit Lines')
plt.ylabel('Monthly Income (in USD)')
plt.xlabel('Loan Amount (in 1000 USD)')
plt.legend(loc="lower right")
plt.show();
```

### Borrower APR relates to Credit Rating and Year

- Bad Credit Ratings have seen a much higher fluctuation in APR.
> Credit Ratings HR, E and D have seen a wide range of fluctuation, rapidly moving up and not coming back down as much. In fact the spread between ratings have increased significantly from 2006 until 2014.

- Good Credit Ratings are have more stable APR.
> While the overall movement of APR over the years has been mostly stable, it is interesting to note that the APR actually reduces during the bad economic period when the average interest rates actually went up.


```python
plt.figure(figsize=(20,10))
sb.pointplot(data=loans, y='BorrowerAPR', x='LoanOriginationYear', ci=90,
           hue='CreditRating', palette='Blues', dodge=-0.5)
plt.legend(title='Credit Rating')

plt.xlabel('Loan Origination Year')
plt.ylabel('Borrower APR')
plt.title('Borrower APR movement through the Years as related to Credit Rating')
plt.show();
```

### Summary: Work on your Credit Rating & Read the Market Conditions

- APR is lower with better Credit Rating.
> We see that Borrower APR very strongly correlates with Credit Rating which in turn depends on an individual and their borrowing and repayment habits. On a personal level, an individual should seek to cultivate habits that increase their credit rating.

- APR varies by the Year.
> What is beyond control though is that market forces and economy tends to move overall interest rates around. This strongly shape what the APR is that one might get based on when they are looking for credit.
